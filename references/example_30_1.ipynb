```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k0 = 0.49 #thermal conductivity (cal/(s*cm*Ùè∞Ø`C))
delta_x = 0.5 #(cm)
delta_t = 0.1 #(s)
length = 10 #(cm)
C = 0.2174 #heat capacity (cal/(g*`C))
dens = 2.7 #density (g/cm^3)
BC1 = 100 #temperature at x=0
BC2 = 50 #temperature at x=10
finaltime = 12 #(s)
k = k0/C/dens #thermal diffusivity (cm^2/s)
lam = k*delta_t/(delta_x**2) 
```

```python
#position we choose
points_number = int(length/delta_x+1)
position = np.linspace(0,length,points_number)
#temperature of all position when t=0
T_point = [0]*points_number
T_point[0] = BC1
T_point[-1] = BC2
T_time = [T_point]
```

```python
def Cal_T(T_point):
    """This function will calculate the temperature at the next timepoint, T_point need to be a list"""
    T_point_new = [BC1]
    for i in range(1,len(T_point)-1):
        new_T = T_point[i]+lam*(T_point[i+1]+T_point[i-1]-2*T_point[i])
        T_point_new.append(new_T)
    T_point_new.append(BC2)
    return T_time.append(T_point_new)
```

```python
#calculate the temperature at all timepoint till t=12(s)
for i in range(int(finaltime/delta_t)):
    Cal_T(T_time[i])
```

```python
for i in [3,6,9,12]:
    plt.plot(position,T_time[int(i/delta_t)],'o-')
plt.xlim(0,10)
plt.legend(('t=3','t=6','t=9','t=12'))
plt.xlabel('Position on rod')
plt.ylabel('T')
```
